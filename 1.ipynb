```python
# Встановлення всіх необхідних бібліотек
!pip install tensorflow
!pip install matplotlib numpy pandas
!pip install scikit-learn
!pip install keras-tuner
!pip install seaborn

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, Dropout, SimpleRNN, Bidirectional
from tensorflow.keras.callbacks import EarlyStopping
import keras_tuner as kt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
import re
import time

# Налаштування для виведення графіків
%matplotlib inline
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 8)

#################### ЗАВДАННЯ 1: ПРОГНОЗУВАННЯ СЛІВ ####################
print("\n" + "="*50)
print("ЗАВДАННЯ 1: ПРОГНОЗУВАННЯ СЛІВ")
print("="*50)

# Завантаження даних
print("Завантаження та підготовка даних...")
try:
    df = pd.read_csv('movie_metadata.csv')
    print("Датасет успішно завантажено!")
    print(f"Кількість записів: {len(df)}")
except FileNotFoundError:
    print("""
    ПОМИЛКА: Файл 'movie_metadata.csv' не знайдено!
    Завантажте датасет з: https://www.kaggle.com/datasets/carolzhangdc/imdb-5000-movie-dataset
    """)
    raise

# Підготовка даних для прогнозування слів
def prepare_text_data(df):
    # Обробка та об'єднання текстових даних
    plot_keywords = df['plot_keywords'].fillna('').astype(str)
    movie_titles = df['movie_title'].fillna('').astype(str)
    director_names = df['director_name'].fillna('').astype(str)

    # Об'єднання всіх текстових даних
    text_data = ' '.join(plot_keywords + ' ' + movie_titles + ' ' + director_names)

    # Очищення тексту
    text_data = re.sub(r'[^\w\s]', '', text_data.lower())
    return text_data

def create_sequences(text, seq_length):
    words = text.split()
    sequences = []
    next_words = []

    for i in range(0, len(words) - seq_length):
        sequences.append(' '.join(words[i:i+seq_length]))
        next_words.append(words[i+seq_length])

    return sequences, next_words

# Параметри для прогнозування слів
SEQUENCE_LENGTH = 5
MAX_WORDS = 10000

# Підготовка даних
text_corpus = prepare_text_data(df)
sequences, next_words = create_sequences(text_corpus, SEQUENCE_LENGTH)

# Токенізація
word_tokenizer = Tokenizer(num_words=MAX_WORDS, oov_token='<OOV>')
word_tokenizer.fit_on_texts(sequences + next_words)
total_words = min(MAX_WORDS, len(word_tokenizer.word_index) + 1)

# Перетворення в послідовності
X_seq = word_tokenizer.texts_to_sequences(sequences)
X_seq = pad_sequences(X_seq)

# Перетворення цільових слів
y_seq = word_tokenizer.texts_to_sequences(next_words)
y_seq = np.array([item[0] for item in y_seq])
y_seq = tf.keras.utils.to_categorical(y_seq, num_classes=total_words)

# Розділення даних
X_train_seq, X_test_seq, y_train_seq, y_test_seq = train_test_split(
    X_seq, y_seq, test_size=0.2, random_state=42
)

print("\nІнформація про дані для прогнозування слів:")
print(f"Загальна кількість слів у словнику: {total_words}")
print(f"Довжина послідовності: {SEQUENCE_LENGTH}")
print(f"Розмір тренувального набору: {X_train_seq.shape}")
print(f"Розмір тестового набору: {X_test_seq.shape}")

# Функція для створення моделі прогнозування слів
def build_word_prediction_model(hp):
    model = Sequential()

    # Tunable parameters
    embedding_dim = hp.Int('embedding_dim', min_value=32, max_value=256, step=32)
    rnn_units = hp.Int('rnn_units', min_value=32, max_value=256, step=32)
    rnn_layers = hp.Int('rnn_layers', min_value=1, max_value=3, step=1)
    dropout_rate = hp.Float('dropout_rate', min_value=0.1, max_value=0.5, step=0.1)

    # Embedding layer
    model.add(Embedding(total_words, embedding_dim, input_length=X_seq.shape[1]))

    # RNN layers
    for i in range(rnn_layers):
        return_sequences = i < rnn_layers - 1
        rnn_type = hp.Choice(f'rnn_type_{i}', ['SimpleRNN', 'LSTM', 'Bidirectional'])

        if rnn_type == 'SimpleRNN':
            model.add(SimpleRNN(rnn_units, return_sequences=return_sequences))
        elif rnn_type == 'LSTM':
            model.add(LSTM(rnn_units, return_sequences=return_sequences))
        else:
            model.add(Bidirectional(LSTM(rnn_units, return_sequences=return_sequences)))

        model.add(Dropout(dropout_rate))

    model.add(Dense(total_words, activation='softmax'))

    # Compilation
    model.compile(
        optimizer='adam',
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )

    return model

# Оптимізація гіперпараметрів
print("\nПочаток оптимізації гіперпараметрів для прогнозування слів...")
word_tuner = kt.RandomSearch(
    build_word_prediction_model,
    objective='val_accuracy',
    max_trials=5,
    directory='word_prediction_tuner',
    project_name='word_prediction'
)

# Навчання з оптимізацією
word_tuner.search(
    X_train_seq,
    y_train_seq,
    epochs=10,
    validation_split=0.2,
    callbacks=[EarlyStopping(patience=2, restore_best_weights=True)]
)
# Отримання найкращої моделі та навчання
best_word_model = word_tuner.get_best_models(num_models=1)[0]
best_word_hp = word_tuner.get_best_hyperparameters(num_trials=1)[0]

print("\nНайкращі гіперпараметри для прогнозування слів:")
for param in ['embedding_dim', 'rnn_units', 'rnn_layers', 'dropout_rate']:
    print(f"{param}: {best_word_hp.get(param)}")

# Навчання найкращої моделі
print("\nНавчання найкращої моделі для прогнозування слів...")
word_history = best_word_model.fit(
    X_train_seq,
    y_train_seq,
    epochs=20,
    validation_split=0.2,
    callbacks=[EarlyStopping(patience=3, restore_best_weights=True)],
    verbose=1
)

#################### ЗАВДАННЯ 2: АНАЛІЗ СЕНТИМЕНТУ ####################
print("\n" + "="*50)
print("ЗАВДАННЯ 2: АНАЛІЗ СЕНТИМЕНТУ")
print("="*50)

# Підготовка даних для аналізу сентименту
def prepare_sentiment_data(df):
    # Створення бінарних міток сентименту на основі оцінок IMDb
    df['sentiment'] = (df['imdb_score'] >= 7.0).astype(int)

    # Об'єднання заголовка та ключових слів
    df['text'] = df['movie_title'].fillna('') + ' ' + df['plot_keywords'].fillna('')
    df['text'] = df['text'].apply(lambda x: re.sub(r'[^\w\s]', '', x.lower()))

    return df[['text', 'sentiment', 'movie_title', 'director_name', 'imdb_score']]

# Параметри для аналізу сентименту
MAX_WORDS_SENTIMENT = 10000
MAX_LEN = 200

# Підготовка даних
sentiment_data = prepare_sentiment_data(df)
texts = sentiment_data['text'].values
labels = sentiment_data['sentiment'].values

# Токенізація для сентименту
sentiment_tokenizer = Tokenizer(num_words=MAX_WORDS_SENTIMENT, oov_token='<OOV>')
sentiment_tokenizer.fit_on_texts(texts)
sequences = sentiment_tokenizer.texts_to_sequences(texts)
X = pad_sequences(sequences, maxlen=MAX_LEN)

# Розділення даних
X_train, X_test, y_train, y_test = train_test_split(X, labels, test_size=0.2, random_state=42)

def create_sentiment_model(model_type):
    model = Sequential()
    model.add(Embedding(MAX_WORDS_SENTIMENT, 100, input_length=MAX_LEN))

    if model_type == 'SimpleRNN':
        model.add(SimpleRNN(64))
    elif model_type == 'LSTM':
        model.add(LSTM(64))
    else:  # Bidirectional
        model.add(Bidirectional(LSTM(64)))

    model.add(Dense(1, activation='sigmoid'))
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

# Навчання моделей
model_types = ['SimpleRNN', 'LSTM', 'Bidirectional']
models = {}
histories = {}
training_times = {}

for model_type in model_types:
    print(f"\nНавчання моделі {model_type}...")
    models[model_type] = create_sentiment_model(model_type)

    start_time = time.time()
    histories[model_type] = models[model_type].fit(
        X_train, y_train,
        epochs=10,
        validation_split=0.2,
        callbacks=[EarlyStopping(patience=2, restore_best_weights=True)],
        verbose=1
    )
    training_times[model_type] = time.time() - start_time

# Функція для візуалізації результатів
def plot_results():
    plt.figure(figsize=(15, 5))

    # Графіки для кожної моделі
    for i, model_type in enumerate(model_types, 1):
        plt.subplot(1, 3, i)
        history = histories[model_type]

        plt.plot(history.history['accuracy'], label='train')
        plt.plot(history.history['val_accuracy'], label='val')
        plt.title(f'{model_type} Accuracy')
        plt.xlabel('Epoch')
        plt.ylabel('Accuracy')
        plt.legend()

    plt.tight_layout()
    plt.show()

# Візуалізація результатів навчання
plot_results()

# Оцінка моделей на тестовому наборі
print("\nРезультати оцінки моделей:")
for model_type in model_types:
    loss, accuracy = models[model_type].evaluate(X_test, y_test, verbose=0)
    print(f"\n{model_type}:")
    print(f"Точність на тестовому наборі: {accuracy:.4f}")
    print(f"Час навчання: {training_times[model_type]:.2f} секунд")

    # Матриця помилок та звіт класифікації
    y_pred = (models[model_type].predict(X_test) > 0.5).astype(int)
    print("\nМатриця помилок:")
    print(confusion_matrix(y_test, y_pred))
    print("\nЗвіт класифікації:")
    print(classification_report(y_test, y_pred))

# Аналіз випадкових рецензій
def analyze_random_reviews(n=10):
    indices = np.random.randint(0, len(X_test), n)

    for idx in indices:
        review_text = sentiment_tokenizer.sequences_to_texts([X_test[idx]])[0]
        true_sentiment = y_test[idx]

        print("\nАналіз рецензії:")
        print(f"Текст: {review_text[:100]}...")
        print(f"Справжній сентимент: {'Позитивний' if true_sentiment == 1 else 'Негативний'}")

        for model_type in model_types:
            pred = models[model_type].predict(X_test[idx:idx+1])[0][0]
            print(f"Передбачення {model_type}: {'Позитивний' if pred > 0.5 else 'Негативний'} "
                  f"(впевненість: {abs(pred-0.5)*2:.2f})")
        print("-"*50)

# Аналіз випадкових рецензій
print("\nАналіз випадкових рецензій:")
analyze_random_reviews()

# Збереження моделей
print("\nЗбереження моделей...")
best_word_model.save('word_prediction_model.keras')
for model_type in model_types:
    models[model_type].save(f'sentiment_{model_type.lower()}.keras')
print("Всі моделі збережено!")

print("\nВиконання завершено!")
```
