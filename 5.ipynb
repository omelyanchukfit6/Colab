```python
# Встановлення необхідних бібліотек
!pip install tensorflow keras matplotlib pandas seaborn

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix

import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau

# Налаштування для відтворюваності результатів
np.random.seed(42)
tf.random.set_seed(42)

# Визначаємо, що файли вже розпаковані
BASE_DIR = '/content/archive/'
print(f"Використовується базова директорія: {BASE_DIR}")

# Завантаження CSV файлу з мітками
df = pd.read_csv(f'{BASE_DIR}sports.csv')
df.head()

# Аналіз розподілу класів
print(f"Загальна кількість зображень: {len(df)}")
print(f"Кількість класів: {df['labels'].nunique()}")
print(f"Набори даних: {df['data set'].unique()}")

# Розподіл зображень за наборами даних
print(df['data set'].value_counts())

# Перегляд розподілу класів
plt.figure(figsize=(12, 8))
sns.countplot(y=df['labels'], order=df['labels'].value_counts().index[:30], hue=df['labels'],
              palette='viridis', legend=False)  # Виправлення попередження
plt.title('Розподіл класів (топ-30)')
plt.xlabel('Кількість зображень')
plt.ylabel('Клас')
plt.tight_layout()
plt.show()

# Розміри зображень для моделі
IMG_SIZE = 224
BATCH_SIZE = 32

# Генератор для тренувального набору з аугментацією
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

# Генератор для валідаційного та тестового наборів без аугментації
valid_datagen = ImageDataGenerator(rescale=1./255)
test_datagen = ImageDataGenerator(rescale=1./255)

# Фільтрація даних за наборами
train_df = df[df['data set'] == 'train']
valid_df = df[df['data set'] == 'valid']
test_df = df[df['data set'] == 'test']

# Генератори даних
train_generator = train_datagen.flow_from_dataframe(
    dataframe=train_df,
    directory=BASE_DIR,
    x_col="filepaths",
    y_col="labels",
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=True
)

valid_generator = valid_datagen.flow_from_dataframe(
    dataframe=valid_df,
    directory=BASE_DIR,
    x_col="filepaths",
    y_col="labels",
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=False
)

test_generator = test_datagen.flow_from_dataframe(
    dataframe=test_df,
    directory=BASE_DIR,
    x_col="filepaths",
    y_col="labels",
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=False
)

# Кількість класів
num_classes = len(train_generator.class_indices)
print(f"Кількість класів: {num_classes}")

# Оскільки попередньо навчена модель несумісна, завжди будемо навчати нову модель
print("Створення нової моделі та навчання з нуля...")

# Використаємо EfficientNetB0 як базову модель для трансферного навчання
from tensorflow.keras.applications import EfficientNetB0

# Створення базової моделі
base_model = EfficientNetB0(weights='imagenet', include_top=False, input_shape=(IMG_SIZE, IMG_SIZE, 3))

# Заморозимо шари базової моделі, щоб не навчати їх спочатку
for layer in base_model.layers:
    layer.trainable = False

# Додаємо наші шари поверх базової моделі
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(1024, activation='relu')(x)
x = Dropout(0.5)(x)
predictions = Dense(num_classes, activation='softmax')(x)

# Створення фінальної моделі
model = Model(inputs=base_model.input, outputs=predictions)

# Компіляція моделі
model.compile(
    optimizer=Adam(learning_rate=0.001),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Відображення структури моделі
model.summary()

# Підготовка колбеків для навчання
checkpoint = ModelCheckpoint(
    'best_model.h5',
    monitor='val_accuracy',
    save_best_only=True,
    mode='max',
    verbose=1
)

early_stopping = EarlyStopping(
    monitor='val_accuracy',
    patience=5,  # Зменшимо patience для швидшого навчання
    restore_best_weights=True,
    mode='max',
    verbose=1
)

reduce_lr = ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.2,
    patience=3,  # Зменшимо patience для швидшого навчання
    min_lr=1e-6,
    mode='min',
    verbose=1
)

callbacks = [checkpoint, early_stopping, reduce_lr]

# Перша фаза навчання - тільки верхні шари (замороженя базова модель)
print("Початок першої фази навчання...")
history = model.fit(
    train_generator,
    epochs=10,  # Зменшимо кількість епох для швидшого навчання
    validation_data=valid_generator,
    callbacks=callbacks,
    verbose=1
)

# Розморозимо деякі верхні шари базової моделі для тонкого налаштування
# Розморозимо останні 30 шарів
print("Розморожування останніх шарів для тонкого налаштування...")
for layer in base_model.layers[-30:]:
    layer.trainable = True

# Скомпілюємо модель з меншою швидкістю навчання для тонкого налаштування
model.compile(
    optimizer=Adam(learning_rate=0.0001),  # Менша швидкість навчання
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Тонке налаштування
print("Початок другої фази навчання (тонке налаштування)...")
fine_tune_history = model.fit(
    train_generator,
    epochs=5,  # Зменшимо кількість епох для швидшого навчання
    validation_data=valid_generator,
    callbacks=callbacks,
    verbose=1
)

# Об'єднання історій навчання
history_combined = {}
for key in history.history.keys():
    history_combined[key] = history.history[key] + fine_tune_history.history[key]

# Візуалізація кривих навчання
plt.figure(figsize=(12, 5))

# Графік точності
plt.subplot(1, 2, 1)
plt.plot(history_combined['accuracy'], label='Навчальна вибірка')
plt.plot(history_combined['val_accuracy'], label='Валідаційна вибірка')
plt.title('Точність моделі')
plt.xlabel('Епоха')
plt.ylabel('Точність')
plt.legend(loc='lower right')
plt.grid(True)

# Графік втрат
plt.subplot(1, 2, 2)
plt.plot(history_combined['loss'], label='Навчальна вибірка')
plt.plot(history_combined['val_loss'], label='Валідаційна вибірка')
plt.title('Функція втрат')
plt.xlabel('Епоха')
plt.ylabel('Значення функції втрат')
plt.legend(loc='upper right')
plt.grid(True)

plt.tight_layout()
plt.show()

# Оцінка моделі на тестовому наборі
print("Оцінка моделі на тестовому наборі...")
evaluation = model.evaluate(test_generator, verbose=1)
print(f"Тестова точність: {evaluation[1]:.4f}")

# Отримання прогнозів для тестового набору
test_generator.reset()
y_pred = model.predict(test_generator, verbose=1)
y_pred_classes = np.argmax(y_pred, axis=1)

# Справжні класи
y_true = test_generator.classes

# Відповідність індексів класам
class_indices = {v: k for k, v in test_generator.class_indices.items()}

# Звіт про класифікацію
print("Звіт про класифікацію:")
print(classification_report(y_true, y_pred_classes, target_names=list(class_indices.values())))

# Візуалізація матриці помилок (для перших 15 класів)
plt.figure(figsize=(15, 12))
conf_mat = confusion_matrix(y_true, y_pred_classes)

# Обмежимо кількість класів для візуалізації (перші 15)
n_classes = 15
conf_mat_subset = conf_mat[:n_classes, :n_classes]
class_names_subset = [class_indices[i] for i in range(n_classes)]

# Нормалізована матриця помилок
conf_mat_norm = conf_mat_subset.astype('float') / conf_mat_subset.sum(axis=1)[:, np.newaxis]

sns.heatmap(conf_mat_norm, annot=True, cmap='Blues', fmt='.2f', xticklabels=class_names_subset, yticklabels=class_names_subset)
plt.title('Нормалізована матриця помилок (перші 15 класів)')
plt.ylabel('Справжні класи')
plt.xlabel('Прогнозовані класи')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Функція для візуалізації Grad-CAM
def make_gradcam_heatmap(img_array, model, last_conv_layer_name, pred_index=None):
    # Модель для виходу останнього згорткового шару і прогнозу
    grad_model = tf.keras.models.Model(
        [model.inputs],
        [model.get_layer(last_conv_layer_name).output, model.output]
    )

    # Записуємо градієнти виходу (прогнозу) відносно виходу останнього згорткового шару
    with tf.GradientTape() as tape:
        last_conv_layer_output, preds = grad_model(img_array)
        if pred_index is None:
            pred_index = tf.argmax(preds[0])
        class_channel = preds[:, pred_index]

    # Градієнт останнього згорткового шару відносно виходу
    grads = tape.gradient(class_channel, last_conv_layer_output)

    # Вектор, де кожен запис - середнє значення градієнта для певного каналу
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))

    # Зважуємо канали за градієнтами
    last_conv_layer_output = last_conv_layer_output[0]
    heatmap = last_conv_layer_output @ pooled_grads[..., tf.newaxis]
    heatmap = tf.squeeze(heatmap)

    # Нормалізація для візуалізації
    heatmap = tf.maximum(heatmap, 0) / tf.math.reduce_max(heatmap)
    return heatmap.numpy()

# Функція для накладання теплової карти на зображення
def save_and_display_gradcam(img, heatmap, alpha=0.4):
    # Масштабуємо теплову карту до розміру зображення
    heatmap = np.uint8(255 * heatmap)
    jet = plt.cm.get_cmap("jet")
    jet_colors = jet(np.arange(256))[:, :3]
    jet_heatmap = jet_colors[heatmap]

    # Створюємо зображення з RGB з теплової карти
    jet_heatmap = tf.keras.preprocessing.image.array_to_img(jet_heatmap)
    jet_heatmap = jet_heatmap.resize((img.shape[1], img.shape[0]))
    jet_heatmap = tf.keras.preprocessing.image.img_to_array(jet_heatmap)

    # Накладання
    superimposed_img = jet_heatmap * alpha + img
    superimposed_img = tf.keras.preprocessing.image.array_to_img(superimposed_img)

    return superimposed_img

# Візуалізуємо активацію на декількох тестових прикладах
test_generator.reset()
num_samples = 5

# Оскільки ми використовуємо свіжу модель EfficientNetB0, ми знаємо назву останнього шару
last_conv_layer_name = "top_activation"  # Для EfficientNetB0

try:
    model.get_layer(last_conv_layer_name)
    print(f"Шар {last_conv_layer_name} знайдено!")

    plt.figure(figsize=(15, num_samples*3))

    for i in range(num_samples):
        # Отримуємо зображення з генератора
        img_batch, label_batch = next(test_generator)
        img = img_batch[0]
        img_array = np.expand_dims(img, axis=0)

        # Отримуємо прогноз
        preds = model.predict(img_array)
        pred_class = np.argmax(preds[0])
        true_class = np.argmax(label_batch[0])

        # Справжній і прогнозований класи
        true_class_name = class_indices[true_class]
        pred_class_name = class_indices[pred_class]
        pred_confidence = preds[0][pred_class]

        # Створюємо Grad-CAM
        heatmap = make_gradcam_heatmap(img_array, model, last_conv_layer_name)

        # Відображаємо оригінальне зображення
        plt.subplot(num_samples, 3, i*3 + 1)
        plt.imshow(img)
        plt.title(f"Справжній: {true_class_name}")
        plt.axis('off')

        # Відображаємо теплову карту
        plt.subplot(num_samples, 3, i*3 + 2)
        plt.imshow(heatmap, cmap='jet')
        plt.title("Теплова карта")
        plt.axis('off')

        # Відображаємо накладання
        superimposed = save_and_display_gradcam(img, heatmap)
        plt.subplot(num_samples, 3, i*3 + 3)
        plt.imshow(superimposed)
        plt.title(f"Прогноз: {pred_class_name} ({pred_confidence:.2f})")
        plt.axis('off')

    plt.tight_layout()
    plt.show()

except Exception as e:
    print(f"Помилка при створенні Grad-CAM: {e}. Візуалізація пропущена.")

# Збереження навченої моделі
model.save('sports_classification_model.h5')
print("Модель збережено як sports_classification_model.h5")

# Висновки
print("""
Висновки:

1. Використання трансферного навчання на основі EfficientNetB0 дозволило побудувати модель
   для класифікації спортивних зображень з досить високою точністю.

2. Стратегія навчання з двома фазами (спочатку заморожені базові шари, потім тонке налаштування)
   показала хороші результати та дозволила ефективно використати попередньо навчені ваги.

3. Візуалізація за допомогою Grad-CAM демонструє, на які частини зображень звертає увагу модель
   при класифікації, що підвищує інтерпретованість моделі.

4. Аналіз матриці помилок і звіту про класифікацію дозволяє виявити проблемні класи,
   що можуть потребувати додаткових даних або спеціальних методів навчання.

5. Трансферне навчання є потужним підходом, який дозволяє швидко створювати ефективні моделі
   класифікації зображень навіть з обмеженим набором даних.
""")
```
